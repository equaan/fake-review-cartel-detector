## Feature Engineering Notes

- `reviewer_features.csv` currently comes from a fast first pass over the first **250,000 cleaned Amazon reviews**.
- The output contains **207,498 reviewers** with normalized reviewer-level features including rating behavior, burstiness, product overlap estimate, text length, and self-similarity.
- `cornell_features.csv` is generated from the downloaded labeled file `fake reviews dataset.csv` and contains TF-IDF features plus label information for supervised classification.
- This subset-based pass is enough to keep the pipeline moving into SVD + DBSCAN without waiting for a full multi-hour Amazon feature build.
